### Importing the modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("/content/wine_data.csv",usecols=[0,1,2],header=None)
```

```python
df
```

```python
df.columns = ["Class label","Alcohol","Malic acid"]
```

```python
df
```

### Kdeplot for independent features

```python
sns.kdeplot(df["Alcohol"])
```

```python
sns.kdeplot(df["Malic acid"])
```

### Scatter Plot between independent features

```python
sns.scatterplot(df["Alcohol"],df["Malic acid"],hue=df["Class label"])
```

### Separate independent and dependent features

```python
X = df.drop("Class label",axis=1)
y = df["Class label"]
```

### Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

### Feature Scaling

```python
sc = MinMaxScaler()
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

### Before Scaling vs After Scaling

```python
fig,ax = plt.subplots(1,2,figsize=(12,5))
ax[0].set_title("Before Scaling")
sns.scatterplot(X_train["Alcohol"],X_train["Malic acid"],hue=df["Class label"],ax=ax[0])
ax[1].set_title("After Scaling")
sns.scatterplot(X_train_scaled["Alcohol"],X_train_scaled["Malic acid"],hue=df["Class label"],ax=ax[1])
```

### Kde Plot for independent features Before Scaling vs After Scaling

```python
fig,ax = plt.subplots(1,2,figsize=(12,5))
ax[0].set_title("Before Scaling")
sns.kdeplot(X_train["Alcohol"],ax=ax[0])
sns.kdeplot(X_train["Malic acid"],ax=ax[0])
ax[1].set_title("After Scaling")
sns.kdeplot(X_train_scaled["Alcohol"],ax=ax[1])
sns.kdeplot(X_train_scaled["Malic acid"],ax=ax[1])
```
